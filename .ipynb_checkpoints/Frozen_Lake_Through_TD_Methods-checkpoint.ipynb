#### Code for testing the implementation

```python
def test(env,agent,num_episodes = 1000):
	R = 0.0
	for episode in range(1,num_episodes+1):

	    if episode%100 == 0:
	        print("\rEpisode {}/{}".format(episode,num_episodes),end = "")
	    # R = 0.0
	    done = False
	    
	    state = env.reset()
	    action = np.argmax(agent.Q[state])

	    while not done:
	        next_state,reward,done,_ = env.step(action)
	        next_action = np.argmax(agent.Q[next_state])
	# 			action = np.random.choice(np.arange(len(probs)),p = probs)
	        # print("State = {},Action = {},Q[s][a] = {}".format(state,action,agent.Q[state]))
	        
	        
	        R += reward
	        state = next_state
	        action = next_action
	        # env.render()
	    # print("-----------")

	print("")
	print(R)
```

```python
def plotting(agent,window_size = 100):
    averaged_returns = np.zeros(len(agent.returns)-window_size+1)
    max_returns = np.zeros(len(agent.returns)-window_size+1)
    min_returns = np.zeros(len(agent.returns)-window_size+1)



    for i in range(len(averaged_returns)):
        averaged_returns[i] = np.mean(agent.returns[i:i+window_size])
        max_returns[i] = np.max(agent.returns[i:i+window_size])
        min_returns[i] = np.min(agent.returns[i:i+window_size])
        
    return (averaged_returns,min_returns,max_returns)

def plot_mean_and_CI(mean, lb, ub, color_mean=None, color_shading=None):
    # plot the shaded range of the confidence intervals
    plt.fill_between(range(mean.shape[0]), ub, lb,
                     color=color_shading, alpha=.5)
    # plot the mean on top
    plt.plot(mean, color_mean)
```

## Implementation Details
- ### Sarsa
    - Sarsa is on-policy algorithm. It implies that it learns the Q-values based on the action performed by the current policy instead of the greedy policy.
    - Update equation of SARSA is
    \begin{equation}
        Q(S_t,A_t) = Q(S_t,A_t) + \alpha \times \big\{R_{t+1} + \gamma Q(S_{t+1},A_{t+1}) - Q(S_t,A_t)\big\}
    \end{equation}
    - Here, the actions are chosen from a epsilon-greedy policy where $\epsilon$ is initially $0.9$ and finally converges to $0.1$ using the decay equation:
    \begin{equation}
        \epsilon_{t} = \epsilon_{min} + \epsilon_{max} \times \exp(-0.01t)
    \end{equation}
    - This is required so as to enable the TD methods to converge. Initially, there is more exploration and eventually it converges to a $\epsilon$ - greedy policy.
- ### Q-Learning
    - Q-Learning is an off policy TD-control policy.It is exactly like SARSA with the only difference being that it doesnot follow a policy to find the next action but rather directly selects the next action based on the current Q(s,a) values.
    - Update equation of Q-Learning is:
    \begin{equation}
        Q(S_t,A_t) = Q(S_t,A_t) + \alpha \times \big\{R_{t+1} + \gamma \max_{a}Q(S_{t+1},a) - Q(S_t,A_t)\big\}
    \end{equation}
    - In order to make the algorithm converge, we manipulate $\epsilon$ as done in SARSA.
    
- ### Expected SARSA
    - Expected SARSA takes the expectation of Q values of every possible action in the current state.
    - Update equation is as follows:
    \begin{equation}
        Q(S_t,A_t) = Q(S_t,A_t) + \alpha \times \big\{R_{t+1} + \gamma \sum_{a} \pi(a|S_{t+1})Q(S_{t+1},a) - Q(S_t,A_t)\big\}
    \end{equation}

## Observations

### Convergence
   - We observe Q-Learning converges much faster as compared to SARSA or Expected SARSA. 
   - In terms of accuracy, SARSA perfoms the best followed by Expected SARSA and then Q-Learning.
   
### Stability
   - All the above TD-Learning algorithms have a problem with convergence and are highly unstable.
   - As Q-Learning uses the max function to estimate Q(s,a) values, it makes highly aggressive updates and is highly unstable.
